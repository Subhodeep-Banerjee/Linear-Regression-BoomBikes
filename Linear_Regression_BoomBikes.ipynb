## Importing the Readme for better understanding of the columns
day.csv have the following fields:

- instant: record index
- dteday : date
- season : season (1:spring, 2:summer, 3:fall, 4:winter)
- yr : year (0: 2018, 1:2019)
- mnth : month ( 1 to 12)
- holiday : weather day is a holiday or not (extracted from http://dchr.dc.gov/page/holiday-schedule)
- weekday : day of the week
- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
+ weathersit : 
    - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
    - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
    - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
    - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
- temp : temperature in Celsius
- atemp: feeling temperature in Celsius
- hum: humidity
- windspeed: wind speed
- casual: count of casual users
- registered: count of registered users
- cnt: count of total rental bikes including both casual and registered

# Data Preparation for modeling

## Creating the dummies

```python
df.info()
```

```python
# Dummy variable creation for month, weekday, weathersit and season variables.

df_month=pd.get_dummies(df.mnth ,drop_first=True)
df_weekdays=pd.get_dummies(df.weekday,drop_first=True)
df_weathersit=pd.get_dummies(df.weathersit,drop_first=True)
df_seasons=pd.get_dummies(df.season,drop_first=True)
```

```python
# Merging  the dataframe, with the dummy variable dataset. 

df_new = pd.concat([df,df_month,df_weekdays,df_weathersit,df_seasons],axis=1)
```

```python
df_new.head()
```

```python
df_new.info()
```

```python
# dropping unnecessary columns as we have already created dummy variable out of it.

df_new.drop(['season','mnth','weekday','weathersit'], axis = 1, inplace = True)
```

```python
# check the head of new dataframe

df_new.head()
```

```python
# check the shape of new dataframe

df_new.shape
```

```python
# check the column info of new dataframe 

df_new.info()
```

## Splitting the data into two groups for train and testing data 

```python
# splitting the dataframe into Train and Test

np.random.seed(0)
df_train, df_test = train_test_split(df_new, train_size = 0.7, random_state = 100)
```

```python
# check the shape of training datatset

df_train.shape
```

```python
# check the shape of testing datatset

df_test.shape
```

### Scaling 

```python
# Using MinMaxScaler to Rescaling the features

scaler = MinMaxScaler()
```

```python
# verifying the head of dataset before scaling.

df_train.head()
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables

num_vars = ['temp','atemp','hum','windspeed','cnt']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
# verifying the head after appying scaling.

df_train.head()
```

```python
# describing the dataset

df_train.describe()
```

```python
# Building the Linear Model

y_train = df_train.pop('cnt')
X_train = df_train
```

## Linear Regression 

```python
# Recursive feature elimination 

lm = LinearRegression()
lm.fit(X_train, y_train)
# Create the RFE object
rfe = RFE(estimator=LinearRegression(), n_features_to_select=15)

# Fit the RFE object to the training data
rfe.fit(X_train, y_train)
```

```python
#List of variables selected in top 15 list

list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# selecting the selected variable via RFE in col list

col = X_train.columns[rfe.support_]
print(col)
```

```python
# checking which columns has been rejected

X_train.columns[~rfe.support_]
```

### Variance Inflation Factor 

```python
# Generic function to calculate VIF of variables

def calculateVIF(df):
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif 
```

```python
# dataframe with RFE selected variables

X_train_rfe = X_train[col]
```

```python
# calculate VIF

calculateVIF(X_train_rfe)
```

 From the variance inflation factor we can see that humidity shows high VIF value.

## Building the liner model

## Model 1

```python
# Building 1st linear regression model

X_train_lm_1 = sm.add_constant(X_train_rfe)
lr_1 = sm.OLS(y_train,X_train_lm_1).fit()
print(lr_1.summary())
```

```python
# As humidity shows high VIF values hence we can drop it
X_train_new = X_train_rfe.drop(['hum'], axis = 1)

# Run the function to calculate VIF for the new model
calculateVIF(X_train_new)
```

## Model 2

```python
# Building 2nd linear regression model

X_train_lm_2 = sm.add_constant(X_train_new)
lr_2 = sm.OLS(y_train,X_train_lm_2).fit()
print(lr_2.summary())
```

```python
# We can drop nov variable as it has high p-value
X_train_new = X_train_new.drop(['holiday'], axis = 1)

# Run the function to calculate VIF for the new model
calculateVIF(X_train_new)
```

Dropping a few variables did not significantly impact the VIF (Variance Inflation Factor) values. Let's proceed further to explore the model and observe for potential improvements.

## Model 3

```python
# Building 3rd linear regression model

X_train_lm_3 = sm.add_constant(X_train_new)
lr_3 = sm.OLS(y_train,X_train_lm_3).fit()
print(lr_3.summary())
```

```python
# We can drop dec variable as it has high p-value
X_train_new = X_train_new.drop(['jan'], axis = 1)

# Run the function to calculate VIF for the new model
calculateVIF(X_train_new)
```

## Model 4

```python
# Building 4th linear regression model

X_train_lm_4 = sm.add_constant(X_train_new)
lr_4 = sm.OLS(y_train,X_train_lm_4).fit()
print(lr_4.summary())
```

```python
# We can drop dec variable as it has high p-value
X_train_new = X_train_new.drop(['july'], axis = 1)

# Run the function to calculate VIF for the new model
calculateVIF(X_train_new)
```

## Model 5

```python
# Building 5th linear regression model

X_train_lm_5 = sm.add_constant(X_train_new)
lr_5 = sm.OLS(y_train,X_train_lm_5).fit()
print(lr_5.summary())
```

### The model lr_5 can be considered, as it exhibits minimal multicollinearity among predictors, and all predictors have significant p-values. The F-Statistic value of 233.6, significantly greater than 1, coupled with the extremely low p-value (1.42e-188, nearly zero), indicates the overall model's significance and reliability in explaining the variance in the data.

## Linearity

```python
# Linear relationship validation using CCPR plot
# Component and component plus residual plot

sm.graphics.plot_ccpr(lr_5, 'temp')
plt.title('CCPR Plot for Temperature')
plt.show()

sm.graphics.plot_ccpr(lr_5, 'windspeed')
plt.title('CCPR Plot for Windspeed')
plt.show()

sm.graphics.plot_ccpr(lr_5, 'sep')
plt.title('CCPR Plot for SEP')
plt.show()
```

The visualizations confirm the presence of linearity in the relationships between the selected features ('temp', 'sep', 'windspeed') and the target variable.

### Homoscedasticity - It refers to the assumption that the variance of the errors (residuals) is constant across all levels of the independent variables.

```python
y_train_pred = lr_5.predict(X_train_lm_5)
residual = y_train - y_train_pred

plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_train, y=residual)
plt.axhline(y=0, color='red', linewidth=1, linestyle='--')
plt.xlabel('Count', fontsize=14)
plt.ylabel('Residual', fontsize=14)
plt.title('Residuals vs Actual Counts', fontsize=16)
plt.show()

```

The scatter plot of residuals against actual counts reveals a random and scattered distribution of points. This lack of any discernible pattern indicates that the model's residuals do not exhibit any systematic behavior concerning the predicted counts.

Durbin-Watson value of final model lr_5 is 2.083, which signifies there is no autocorrelation.
